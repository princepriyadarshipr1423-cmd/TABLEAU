# ****CALCULATED FIELDS, PARAMETERS & LOD EXPRESSIONS** ✔️**

**Question 1 :
What is a Calculated Field and why is it important in business dashboards?**

**Ans-**

**Why important in dashboards? 📊**

- Enables custom metrics (profit, growth, ratios)
- Helps in data transformation without changing source data
- Supports better analysis & insights
- Makes dashboards dynamic and flexible

**Question 2 :
Why can’t parameters change visuals directly in Tableau**.

**Ans-**

**Parameters cannot change visuals directly because they are independent of the data model.**

Key point ⚙️

* They only store a value
- They affect visuals only when used in a calculated field, filter, or reference line

**Question 3 :
When should EXCLUDE LOD be used**

**Ans-**

**Use EXCLUDE LOD when you want to remove a dimension from the current view’s level of detail.**

 **When to use? 📊**


- To calculate values at a higher aggregation level
When you want results independent of a specific dimension in the view

**Example **  
View: Sales by Region + Category
EXCLUDE Category → gives total sales per Region only




**Question 4 :
Difference between FIXED and INCLUDE LOD expressions**

**Ans-**

| Feature            | FIXED                       | INCLUDE                       |
| ------------------ | --------------------------- | ----------------------------- |
| Dependency on view | ❌ Independent               | ✅ Depends on view             |
| Behavior           | Ignores dimensions in view  | Adds extra dimensions to view |
| Control            | Strict (fixed level)        | Flexible (extends level)      |
| Filters            | Ignores non-context filters | Respects filters              |
